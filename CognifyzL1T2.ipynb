```python
import pandas as pd
df = pd.read_csv('/content/Dataset .csv')
```

```python
 City_counts = df['City'].value_counts()
 most_Restaurants_City = City_counts.idxmax()
 print(most_Restaurants_City)
```

```python
City_ratings = df.groupby('City')['Aggregate rating'].mean()
print(City_ratings)

# convert the dataframe to a string and print it
print(df.to_string())
```

```python
highest_rating_city = City_ratings.idxmax()
print(highest_rating_city)
```
