```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Dataset .csv')
```

```python
df.info()
```

```python
highest_votes = df.nlargest(1, 'Votes')['Restaurant Name']
lowest_votes = df.nsmallest(1, 'Votes')['Restaurant Name']
print('Restaurants with the highest number of votes:')
print(highest_votes)


print('Restaurants with the lowest number of votes:')
print(lowest_votes)
```

```python
sns.regplot(x='Votes', y='Aggregate rating', data=df)
plt.show()

print('Correlation between votes and Aggregate rating:')
print(np.corrcoef(df['Votes'], df['Aggregate rating'])[0, 1])
```
