```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Dataset .csv')
```

```python
df.info()
```

```python
"""
Analyze if there is a relationship between the
price range and the availability of online
delivery and table booking.
Determine if higher-priced restaurants are
more likely to offer these services.


Price range ,  Has Online delivery  ,   Has Table booking
"""
```

```python
sns.countplot(x='Price range', hue='Has Online delivery', data=df)
plt.show()

online_delivery_table = pd.crosstab(df['Price range'], df['Has Online delivery'])
print(online_delivery_table)
```

```python
Has_Online_delivery_table = pd.crosstab(df['Price range'], df['Has Online delivery'])
```

```python
Has_Table_booking_table = pd.crosstab(df['Price range'], df['Has Table booking'])
```

```python
online_delivery_percentage = Has_Online_delivery_table.apply(lambda x: x / x.sum() * 100, axis=1)
print(online_delivery_percentage)

table_booking_percentage = Has_Table_booking_table.apply(lambda x: x / x.sum() * 100, axis=1)
print(table_booking_percentage)
```

# Based on the results that we got from analysis is :

-->There is a negative relationship between the price range and the availability of online delivery.
This means that as the price range increases, the percentage of restaurants that offer online delivery decreases, and vice versa.
This could be because higher-priced restaurants are more likely to have a dine-in experience and less likely to cater to online orders.


-->There is a positive relationship between the price range and the availability of table booking.
This means that as the price range increases, the percentage of restaurants that offer table booking increases, and vice versa.
This could be because higher-priced restaurants are more likely to have a high demand and less capacity, and therefore require reservations in advance.


-->Higher-priced restaurants are **not** more likely to offer online delivery. In fact, the percentage of restaurants that offer online delivery decreases as the price range increases.
Only 9.04% of the restaurants in the highest price range  offer online delivery, compared to 15.77%, 41.31%, and 29.19% of the restaurants in the lower price ranges .

-->Higher-priced restaurants are **more** likely to offer table booking. The percentage of restaurants that offer table booking increases as the price range increases.
Almost half of the restaurants in the highest price range  offer table booking, compared to 0.02%, 7.68%, and 45.74% of the restaurants in the lower price ranges .
