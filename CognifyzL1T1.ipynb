```python
import pandas as pd
```

```python
# Load the dataset into a pandas dataframe
df = pd.read_csv('/Dataset .csv')
```

```python
#To find the most common cuisines, you can use the value_counts() method on the cuisine column. This will return a series with the counts of each unique value in the column. You can use the head(3) method to get the top three values. You can use the following code
Cuisines_counts = df['Cuisines'].value_counts()
top_Cuisines = Cuisines_counts.head(3)
print(top_Cuisines)
```

```python
#To calculate the percentage of restaurants that serve each of the top cuisines, you can divide the counts by the total number of restaurants in the dataset. You can use the len() function to get the number of rows in the dataframe. You can use the following code:
total_Restaurants = len(df)
top_Cuisines_percentage = top_Cuisines / total_Restaurants * 100
print(top_Cuisines_percentage)
```
